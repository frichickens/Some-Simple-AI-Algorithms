```python
# One-hot encode Gender and Smokeing_history
```

```python
data = data.join(pd.get_dummies(data.gender, dtype=int)).drop(['gender'],axis = 1)
data = data.join(pd.get_dummies(data.smoking_history, dtype=int)).drop(['smoking_history'],axis = 1)
```

```python
# Normalize data
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
for col in data:
    data[col] = scaler.fit_transform(data[[col]])
```

```python
# Shuffle data to create more randomness
data = data.sample(frac = 1)
```

```python
# Splitting data into train data and test data
```

```python
train_data = data.iloc[:int(m*80/100),:]
test_data = data.iloc[int(m*80/100):,:]
```

```python
X_train = np.array(train_data.drop(['diabetes'],axis = 1)).T
Y_train = np.array(train_data['diabetes'])
```

```python
def init_params():
    W1 = np.random.randn(8,15)
    b1 = np.random.randn(8,1)
    W2 = np.random.randn(2,8)
    b2 = np.random.randn(2,1)
    return W1, b1, W2, b2

def ReLU(Z):
    return np.maximum(Z,0)
    
def Stable_Softmax(Z):
    c = np.max(Z)
    A = np.exp(Z-c) / sum(np.exp(Z-c))
    return A

def deriv_ReLU(Z):
    return Z > 0

def one_hot(Y):
    a = np.array(Y)
    b = np.where(a<1, 1, 0)
    one_hot_Y = np.array([b,a])
    return one_hot_Y

def foward_prob(W1, b1, W2, b2, X):
    Z1 = W1.dot(X) + b1
    A1 = ReLU(Z1)
    Z2 = W2.dot(A1) + b2
    A2 = Stable_Softmax(Z2)
    return Z1, A1, Z2, A2

def back_prob(Z1, A1, Z2, A2, W1, W2, X, Y):
    # derivative of Loss-Entropy w.r.t Z2, W2, b2, Z1, W1, b1
    dZ2 = A2 - one_hot(Y)
    dW2 = 1 / m * dZ2.dot(A1.T)
    db2 = 1 / m * np.sum(dZ2)
    dZ1 = W2.T.dot(dZ2) * deriv_ReLU(Z1)
    dW1 = 1 / m * dZ1.dot(X.T)
    db1 = 1 / m * np.sum(dZ1)
    return dW1, db1, dW2, db2

def update_params(W1, b1, W2, b2, dW1, db1, dW2, db2, learning_rate):
    # gradient descent
    W1 = W1 - learning_rate * dW1
    b1 = b1 - learning_rate * db1
    W2 = W2 - learning_rate * dW2
    b2 = b2 - learning_rate * db2
    return W1, b1, W2, b2
```

```python
def get_predictions(A2):
    return np.argmax(A2,0)

def get_accuracy(predictions, Y):
    print(predictions, Y)
    return np.sum(predictions == Y) / Y.size

def gradient_descent(X, Y, learning_rate, iter):
    W1, b1, W2, b2 = init_params()
    for i in range(iter):
        Z1, A1, Z2, A2 = foward_prob(W1, b1, W2, b2, X)
        dW1, db1, dW2, db2 = back_prob(Z1, A1, Z2, A2, W1, W2, X, Y)
        W1, b1, W2, b2 = update_params(W1, b1, W2, b2, dW1, db1, dW2, db2, learning_rate)
        if i % 10 == 0:
            print("Iteration: ", i)
            predictions = get_predictions(A2)
            print(get_accuracy(predictions, Y))
    return W1, b1, W2, b2
```

```python
W1, b1, W2, b2 = gradient_descent(X_train, Y_train, 0.10, 500)
```

```python
def make_predictions(X, W1, b1, W2, b2):
    Z1, A1, Z2, A2 = foward_prob(W1, b1, W2, b2, X)
    predictions = get_predictions(A2)
    return predictions

X_test = np.array(test_data.drop(['diabetes'],axis = 1)).T
Y_test = np.array(test_data['diabetes'])

test_predictions = make_predictions(X_test, W1, b1, W2, b2)
get_accuracy(test_predictions, Y_test)
```
